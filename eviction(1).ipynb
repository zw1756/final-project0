```python
# 2
df['count'] =1
df['year'] = df['Executed Date'].map(lambda x:x[-4:])
df
```

```python
df1 = df.groupby('year')['count'].sum().to_frame()
df1
```

```python
year_largest = df1[df1['count']==df1['count'].max()].index.values[0]
year_smallest = df1[df1['count']==df1['count'].min()].index.values[0]

print("the largest number of evictions:" + str(year_largest))
print("the smallest number of evictions:" + str(year_smallest))
```

```python
# 3
# 求Residential的数据
r = len(df[df['Residential/Commercial']=="Residential"])
r
```

```python
# 求Commercial的数据
c = len(df[df['Residential/Commercial']=="Commercial"])
c
```

```python
# 求比例
ratio = c / r
ratio
```

```python
#6 
from shapely.geometry import Point 
import geopandas as gpd

geometryO = [Point(xy) for xy in zip(df['Latitude'],df['Longitude'])]
gdf_O = gpd.GeoDataFrame(df, crs="EPSG:4326", geometry=geometryO)
gdf_O.plot(color='red',legend=True,figsize=(60,30),markersize=0.1)
```

```python
# 7 Community Board
base = gdf_O.plot(figsize=(60,30), linewidth=5, edgecolor='dimgray', facecolor='lightgrey',alpha = 0.5,legend = True)
gdf_O.plot(ax = base, column='Community Board', cmap='plasma', figsize=(120,60),
                     linewidth=2, edgecolor='white', facecolor='lightgrey', legend = True,vmax=200)
```
